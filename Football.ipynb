Visualization

```python
#How has the number of goals scored changed over the years?
import pandas as pd
import matplotlib.pyplot as plt
results['date'] = pd.to_datetime(results['date'])
results['year'] = results['date'].dt.year
results['total_goals']=results['away_score']+results['home_score']
results.groupby('year')['total_goals'].sum().plot(kind='line',figsize=(12,6),title='Number of goals scored by year')
plt.xlabel('Year')
plt.ylabel('Number of goals')
plt.grid(True)
plt.show()
```

```python
#Which are the top 10 national teams with the most goals scored??
home_goals = results.groupby('home_team')['home_score'].sum()
away_goals = results.groupby('away_team')['away_score'].sum()
total_goals = home_goals.add(away_goals, fill_value=0)
total_goals.sort_values(ascending=False).head(10).plot(kind='barh',figsize=(12,6),title='the top 10 national teams with the most goals')
plt.xlabel('Number of Goals')
plt.ylabel('Team')
plt.show()
```

```python
#How are goals distributed by minute?
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.histplot(goalscores['minute'], bins=30)
plt.title('Goals Distribution by Minute')
plt.xlabel('Minute')
plt.ylabel('Number of Goals')
plt.show()
```

```python
#What is the ratio of penalty to non-penalty goals?
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(goalscores,x='penalty')
plt.title('the ratio of penalty to non-penalty goals')
plt.xlabel('Penalty')
plt.ylabel('Number of goals')
plt.show()
```

```python
#What is the yearly trend of own goals?
import matplotlib.pyplot as plt
goalscores['date'] = pd.to_datetime(goalscores['date'])
goalscores['year'] = goalscores['date'].dt.year
own_goals_yearly = goalscores.groupby('year')['own_goal'].sum()
own_goals_yearly.plot(kind='line',figsize=(16,8),marker='o',title='the yearly trend of own goals')
plt.xlabel('Year')
plt.ylabel('Number of Own goals')
plt.grid(True)
plt.show()

```

```python
#Which are the top 15 countries with the most matches played?
import matplotlib.pyplot as plt
top_countries=results['country'].value_counts().head(15)
top_countries.plot(kind='bar',figsize=(12,6),title='the top 15 countries with the most matches')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.xlabel('Countries')
plt.ylabel('Number of matches')
plt.show()
```

```python
#How are total goals distributed in matches?
import matplotlib.pyplot as plt
import seaborn as sns
total_goals=results['home_score']+results['away_score']
sns.histplot(total_goals,bins=20)
plt.title('Distribution of total goals in matches')
plt.xlabel('Goals')
plt.ylabel('Number of Matches')
plt.show()
```

```python
import matplotlib.pyplot as plt
top_tournaments = results['tournament'].value_counts().head(10)
top_tournaments.plot(kind='bar',figsize=(12,6),title='Top 10 tournaments by number of matches played')
plt.xlabel('Tournaments')
plt.ylabel('Number of matches')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
#How do the top 20 goal scorers compare?
top_scorers = goalscores.groupby('scorer').size().sort_values(ascending=False).head(20)
top_scorers.plot(kind='barh',figsize=(12,6),title='Top 20 players with the most goals scored')
plt.xlabel('Number of golas')
plt.ylabel('Player')
plt.show()
```
