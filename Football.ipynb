Visualization

```python
#İllər üzrə vurulan qolların sayı necə dəyişib?
import pandas as pd
import matplotlib.pyplot as plt
results['date'] = pd.to_datetime(results['date'])
results['year'] = results['date'].dt.year
results['total_goals']=results['away_score']+results['home_score']
results.groupby('year')['total_goals'].sum().plot(kind='line',figsize=(12,6),title='Iller üzrə vurulan qollarin sayi')
plt.xlabel('il')
plt.ylabel('Qollarin sayi')
plt.grid(True)
plt.show()
```

```python
#Ən çox qol vuran 10 milli komanda hansılardır?
home_goals = results.groupby('home_team')['home_score'].sum()
away_goals = results.groupby('away_team')['away_score'].sum()
total_goals = home_goals.add(away_goals, fill_value=0)
total_goals.sort_values(ascending=False).head(10).plot(kind='barh',figsize=(12,6),title='Ən çox qol vuran 10 milli komanda')
plt.xlabel('Qol sayı')
plt.ylabel('Komanda')
plt.show()
```

```python
#Qolların dəqiqələr üzrə paylanması necədir?
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.histplot(goalscores['minute'], bins=30)
plt.title('Qolların dəqiqələr üzrə paylanması')
plt.xlabel('Minute')
plt.ylabel('Number of Goals')
plt.show()
```

```python
#Penalti və qeyri-penalti qollarının nisbəti nə qədərdir?
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(goalscores,x='penalty')
plt.title('Penalti və qeyri-penalti qolları')
plt.xlabel('Penalti')
plt.ylabel('Qol sayı')
plt.show()
```

```python
#Avtoqol (own goal) hadisələrinin illər üzrə trendi necədir?
import matplotlib.pyplot as plt
goalscores['date'] = pd.to_datetime(goalscores['date'])
goalscores['year'] = goalscores['date'].dt.year
own_goals_yearly = goalscores.groupby('year')['own_goal'].sum()
own_goals_yearly.plot(kind='line',figsize=(16,8),marker='o',title='İllər üzrə avtoqolların sayı')
plt.xlabel('İl')
plt.ylabel('Avtoqol sayı')
plt.grid(True)
plt.show()

```

```python
#Ən çox oyun keçirilən 15 ölkə hansılardır?
import matplotlib.pyplot as plt
top_countries=results['country'].value_counts().head(15)
top_countries.plot(kind='bar',figsize=(12,6),title='Ən çox oyun keçirilən 15 ölkə')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.xlabel('Ölkələr')
plt.ylabel('oyun sayı')
plt.show()
```

```python
#Matçlarda vurulan ümumi qolların paylanması necədir?
import matplotlib.pyplot as plt
import seaborn as sns
total_goals=results['home_score']+results['away_score']
sns.histplot(total_goals,bins=20)
plt.title('Matçlarda vurulan ümumi qolların paylanması')
plt.xlabel('Qollar')
plt.ylabel('Matç sayı')
plt.show()
```

```python
import matplotlib.pyplot as plt
top_tournaments = results['tournament'].value_counts().head(10)
top_tournaments.plot(kind='bar',figsize=(12,6),title='Turnirlər üzrə oyun sayı (Top 10)')
plt.xlabel('Turnir')
plt.ylabel('Oyun sayı')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
#Ən çox qol vuran 20 futbolçunun nəticələri necə müqayisə olunur?
top_scorers = goalscores.groupby('scorer').size().sort_values(ascending=False).head(20)
top_scorers.plot(kind='barh',figsize=(12,6),title='Ən çox qol vuran 20 futbolçu')
plt.xlabel('Qol sayı')
plt.ylabel('Futbolçu')
plt.show()
```
